```python
import nltk
```

```python
#creating a set of words from alice in wonderland
aliceWordSet = nltk.corpus.gutenberg.words('carroll-alice.txt')
```

```python
#Frequency Distribution
aliceWordSetFd = nltk.FreqDist(aliceWordSet)
```

```python
aliceWordSetFd
```

```python
#lookup for the word rabbit
aliceWordSetFd['Rabbit']
```

```python
#Most Common 15 word
aliceWordSetFd.most_common(15)
```

```python
#Word used only a once
aliceWordSetFd.hapaxes()
```
