```python
import nltk
import random
import matplotlib
import matplotlib.pyplot as plot
```

```python
from nltk.corpus import names
```

```python
names.fileids()
```

```python
matplotlib.style.use("ggplot")
%matplotlib inline
```

```python
#conditional frequency distribution
nameCFd = nltk.ConditionalFreqDist((file, name[-2:]) for file in names.fileids() for name in names.words(file))
```

```python
plot.figure(figsize=(50,10))
nameCFd.plot()
```

```python
def lastTwoLetter(name):
    return {"pair " : name[-2:]}
```

```python
lastTwoLetter("katy")
```

```python
nameList = [(name, "male") for name in names.words("male.txt")] + [(name, "female") for name in names.words("female.txt")]
```

```python
nameList[:10]
```

```python
nameList[-10:]
```

```python
#Randomise
random.shuffle(nameList)
```

```python
nameList[:10]
```

```python
features = [(lastTwoLetter(name), gender) for (name, gender) in nameList]
```

```python
features[:10]
```

```python
#finding middle point of list features
len(features)/2
```

```python
traningSet = features[:3972]
testSet    = features[3972:]
```

```python
#Naive Bayes classifier to train our model
classifier = nltk.NaiveBayesClassifier.train(traningSet)
```

```python
#we can check our classfier
#load male list
maleNamesList = names.words("male.txt")
```

```python
#check whether name sabin is in male name list provided by nltk
'sabin' in maleNamesList
```

```python
#Let's find out what our classifer thinks of the name
classifier.classify(lastTwoLetter('sabin'))
```

```python
#let's check how well our model did
nltk.classify.accuracy(classifier, testSet)
```
